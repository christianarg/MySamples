```python
import matplotlib.pyplot as plt
import numpy as np

origin = np.array([0, 0])

def plot(v):
    npv = np.array(v)
    plt.quiver(*origin, *npv, scale=1, scale_units='xy', angles='xy', color='b')

plot([100, 10])    
plot([10, 100])    


plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.grid()
plt.show()
```
